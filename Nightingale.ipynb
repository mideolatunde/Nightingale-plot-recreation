```python
#Importing necessary packages
import numpy as np
import pandas as pd
from plotnine import *
from janitor import *
import matplotlib.pyplot as plt
from lets_plot import *
LetsPlot.setup_html(no_js=True)
```

```python
#Loading the 1854-1855 dataset and checking the entries
data_1854_1855 = pd.read_excel('raw_night_1854_1855.xlsx')
data_1854_1855.head(12)
```

```python
#Loading the 1855-1856 dataset and checking the entries
data_1855_1856 = pd.read_excel('raw_night_1855_1856.xlsx')
data_1855_1856.head(12)
```

```python
#Checking Information about the dataset
print(data_1854_1855.info())

print(data_1855_1856.info())
```

### Data Cleaning

```python
#Transforming the data from wide to long
data_1854_1855 = data_1854_1855.pivot_longer(index = ['Month', 'Avg Army Size'],
                                             names_to = 'Cause of Death', 
                                             values_to = 'Number')
data_1854_1855

# Creating a new DataFrame with the Month Column splitted into Month and Year
data1 = pd.DataFrame(data_1854_1855['Month'].str.split(' ', expand= True))
# Renaming the new columns created 
data1 = data1.rename(columns={0: 'Month_2', 1: 'Year'})

# Joining the previous DataFrame with the new DataFrame
data_54 = pd.concat([data_1854_1855, data1], axis=1)

#Replacing the Entries to their full name
data_54['Month_2'] = data_54['Month_2'].replace(['Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec', 'Jan', 'Feb', 'Mar'], ['APRIL', 'MAY', 'JUNE', 'JULY', 'AUGUST', 'SEPTEMBER', 'OCTOBER', 'NOVEMBER', 'DECEMBER', 'JANUARY', 'FEBRUARY', 'MARCH'])
data_54['Number'] = data_54['Number'].astype('int')
data_54['Cause of Death'] = data_54['Cause of Death'].str.replace('Disease Deaths', 'Diseases')
data_54
```

```python
#Checking the Statistical description of the datasets
print(data_1854_1855.describe())

print(data_1855_1856.describe())
```

### Visualising the Coxcomb plot for year 1854 - 1855

```python
# Creating the radius column which is the square root for better visualization
data_54['radius'] = np.sqrt(data_54['Number'])

#Transforming the Cause of Death column to categorical column
data_54['Cause of Death'] = pd.Categorical(data_54['Cause of Death'],
              categories = ["Wounds", "Other Causes", "Diseases"],
              ordered = True)
```

```python
p1 = (ggplot(data_54, aes('Month_2', 'radius', fill = 'Cause of Death')) 
 + geom_bar(stat='identity', width = 1, alpha = 0.8, color='grey') 
 + coord_polar(direction=1, start= -np.pi/2.4)
 + theme_minimal()
 + scale_fill_manual(values = ["#fb9a99", "grey30", "skyblue"]) 
+ labs(title = "Chart showing the Causes of Death in the Crimean War \nfrom APRIL 1854 - MARCH 1855") 
+ theme(axis_title_y = element_blank(),
        axis_title_x = element_blank(),
        axis_text_y = element_blank(),
        axis_ticks = element_blank(),
        plot_title = element_text(size=15, face='bold', hjust = 0.5),
        panel_grid = element_blank(),
        panel_border = element_blank(),
        legend_position='bottom')
+  ggsize(900, 600))
p1
```

#### 1855 - 1856 (Data Cleaning)

```python
#Transforming the data from wide to long
data_1855_1856 = data_1855_1856.pivot_longer(index = ['Month', 'Avg Army Size'],
                                             names_to = 'Cause of Death', 
                                             values_to = 'Number')
data_1855_1856

# Creating a new DataFrame with the Month Column splitted into Month and Year
data1 = pd.DataFrame(data_1855_1856['Month'].str.split(' ', expand= True))
# Renaming the new columns created 
data1 = data1.rename(columns={0: 'Month_2', 1: 'Year'})

# Joining the previous DataFrame with the new DataFrame
data_55 = pd.concat([data_1855_1856, data1], axis=1)

#Replacing the Entries to their full name
data_55['Month_2'] = data_55['Month_2'].replace(['Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec', 'Jan', 'Feb', 'Mar'], ['APRIL', 'MAY', 'JUNE', 'JULY', 'AUGUST', 'SEPTEMBER', 'OCTOBER', 'NOVEMBER', 'DECEMBER', 'JANUARY', 'FEBRUARY', 'MARCH'])
data_55['Number'] = data_55['Number'].astype('int')
data_55['Cause of Death'] = data_55['Cause of Death'].str.replace('Disease Deaths', 'Diseases')
data_55
```

### Visualising the Coxcomb plot for year 1855 - 1856

```python
data_55['radius'] = np.sqrt(data_55['Number'])
data_55['Cause of Death'] = pd.Categorical(data_55['Cause of Death'], 
                                           categories=['Wounds', 'Other Causes', 'Diseases'],
                                           ordered = True)
data_55
```

```python
p2 = (ggplot(data_55, aes('Month_2', 'radius', fill = 'Cause of Death')) 
 + geom_bar(stat='identity', width = 1, alpha = 0.6, color = 'grey') 
 + coord_polar(direction=1, start= -np.pi/2.4)
 + theme_minimal()
 + scale_fill_manual(values = ["#fb9a99", "grey30", "skyblue"]) 
+ labs(title = "Chart showing the Causes of Death in the Crimean War \nfrom APRIL 1855 - MARCH 1856") 
+ theme(axis_title_y = element_blank(),
        axis_title_x = element_blank(),
        axis_text_y = element_blank(),
        axis_ticks = element_blank(),
        plot_title = element_text(size=15, face='bold', hjust = 0.5),
        panel_grid = element_blank(),
        panel_border = element_blank(),
        legend_position='bottom')
+  ggsize(900, 600))
p2
```
